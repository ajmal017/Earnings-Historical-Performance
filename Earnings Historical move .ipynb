```python
import pandas as pd
import numpy as np 
import yfinance as yf
from datetime import datetime
from datetime import timedelta
from yahoo_earnings_calendar import YahooEarningsCalendar
import dateutil.parser
```

```python
def getAverage(twtr):
    print("day1: "+str(round(twtr['day1'].mean(),2)))
    print("day2: "+str(round(twtr['day2'].mean(),2)))
    print("day5: "+str(round(twtr['day5'].mean(),2)))
    print("day10: "+str(round(twtr['day10'].mean(),2)))
    print("day20: "+str(round(twtr['day20'].mean(),2)))
    print("day30: "+str(round(twtr['day30'].mean(),2)))
    print("day60: "+str(round(twtr['day60'].mean(),2)))
    print("day90: "+str(round(twtr['day90'].mean(),2)))
    print("day365: "+str(round(twtr['day365'].mean(),2)))
    
def earningsPerformance(ticker):
    #Company Specific earnings 
    # setting the parameters 
    TICKER = ticker
    DAYS_AHEAD = 180

    # setting the dates
    #start_date = datetime.now().date()
    #end_date = (datetime.now().date() + timedelta(days=DAYS_AHEAD))

    start_date = datetime.now().date()- timedelta(days=2500)
    end_date = (datetime.now().date())+timedelta(days= 90)

    # downloading the earnings calendar
    yec = YahooEarningsCalendar()
    earnings_list = yec.get_earnings_of(TICKER)
    earnings_df = pd.DataFrame(earnings_list)

    # extracting the date from the string and filtering for the period of interest
    earnings_df['report_date'] = earnings_df['startdatetime'].apply(lambda x: dateutil.parser.isoparse(x).date())
    earnings_df = earnings_df.loc[earnings_df['report_date'].between(start_date, end_date)] \
                             .sort_values('report_date')

    twtr = earnings_df
    twtr = twtr[twtr['startdatetimetype']=='TNS']
    twtr = twtr[twtr['epsactual'].notnull()]
    twtr = twtr[['ticker','epsestimate','epsactual','epssurprisepct','report_date']]
    #dates after 1,2,5,10,20,60,90
    twtr["day1"] = np.nan
    twtr["day2"] = np.nan
    twtr["day5"] = np.nan
    twtr["day10"] = np.nan
    twtr["day20"] = np.nan
    twtr["day30"] = np.nan
    twtr["day60"] = np.nan
    twtr["day90"] = np.nan
    twtr["day365"] = np.nan
    price = yf.download(TICKER,threads= False, start = '2010-09-21', end = '2021-05-09', interval = '1d')
    #start_date.strftime("%Y-%m-%d")
    for ind,row in twtr.iterrows():
        ilocation = 0 
        curDate = str(row.loc["report_date"])
        #curDate = curDate+" 00:00:00"
        try:
            currentPrice = float(price[price.index == curDate]['Adj Close'][0])
        except:
            pass
        try:
            ilocation = price.index.get_loc(price[price.index == curDate].iloc[0].name)
        except:
            print("location issue")
            pass
        try:
            day1 = price.iloc[ilocation+1]['Adj Close']
        except:
            
            day1 = np.nan
            pass
        try:
            day2 = price.iloc[ilocation+2]['Adj Close']
        except:
            
            day2 = np.nan
            pass
        try:
            day5 = price.iloc[ilocation+5]['Adj Close']
        except:
            
            day5 = np.nan
            pass
        try:
            day10 = price.iloc[ilocation+10]['Adj Close']
        except:
            
            day10 = np.nan
            pass
        try:
            day20= price.iloc[ilocation+20]['Adj Close']
        except:
            
            day20 = np.nan
            pass
        try:
            day30 = price.iloc[ilocation+30]['Adj Close']
        except:
            
            day30 = np.nan
            pass
        try:
            day60 = price.iloc[ilocation+60]['Adj Close']
        except:
            
            day60 = np.nan
            pass
        try:
            day90 = price.iloc[ilocation+90]['Adj Close']
        except:
            
            day90 = np.nan
            pass
        try:
            day365= price.iloc[ilocation+365]['Adj Close']
        except:
            
            day365 = np.nan
            pass


        try:
            twtr.loc[ind,"day1"]= ((day1-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day2'] = ((day2-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day5'] = ((day5-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day10'] = ((day10-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day20'] = ((day20-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day30'] = ((day30-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day60'] = ((day60-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day90'] = ((day90-currentPrice)/currentPrice)*100
        except:
            pass
        try:
            twtr.loc[ind,'day365']=((day365-currentPrice)/currentPrice)*100
        except:
            pass

    getAverage(twtr)
    
    return twtr



```
